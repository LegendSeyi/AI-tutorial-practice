## ONE HOT ENCODER

One hot encoder is really good when we are working on a categorical data

```python
#import a categorical Data

d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df = pd.DataFrame(d)
```

```python
df.head()
```

```python
df['city'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# handle_unknown='ignore' - to ignore unknown values
# sparse_output=False  - to have our output in pandas dataframe.
#.set_output(transform='pandas') - for a pandas output

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
# transform to One hot encoder

ohetransform = ohe.fit_transform(df[['city']])
```

```python
ohetransform
```

what this did in simple term is that for the City that is suppose to be on the row, we will have the 1.0 value, and for the cities that are not specific row, they will have the Value zero.


that means that the categories were converted into columns and has the value of 1.0 or 0.0 depending if they appear on a row or not.

#### we want to Concat, that means that we want to add the Ohe dataframe (after we have done the one hot encoder to the city category) to the original dataframe, then we drop the city category. make the data more numerical

```python
df = pd.concat([df, ohetransform], axis=1).drop(columns=['city'])
```

```python
df.head()
```

### Ordinal Encoder

this simply the changing of categorical data to Numbers

```python
## create a dataset to demonstrate the code and process
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
d = pd.DataFrame(d)
```

```python
d.head(10)
```

We can observe that the string columnshave two type of data
one is Nominal (City), the other one is Ordinal (size)

```python
d['size'].unique()
```

```python
# we assign this as a dictionary like format
size = ['Small', 'Medium', 'Large']
```

```python
# we called the ordinal encoding preprosess library from sklearn
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder(categories=[size])
```

```python
enc
```

```python
# we then use our fit transform

enc.fit_transform(d[['size']])
```

```python
# so we apply this to the dataset
d['size'] = enc.fit_transform(d[['size']])
```

```python
d
```

this is the ordinal encoding is done using sklearn, this an important preprocessing method because some ML algorithm do not recognize string values
